```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plp
import seaborn as sn
import datetime as dt
```

```python
data = pd.read_csv('C:\\Users\\Saumya Jain\\OneDrive\\Desktop\\dataset\\USvideos_modified.csv')
```

```python
data['publish_date'] = pd.to_datetime(data['publish_date'])
```

```python
data.head()
```

```python
data['publish_month']=data['publish_date'].dt.month
data['publish_day']=data['publish_date'].dt.day
data.head()
```

```python
print(sorted(data['category_id'].unique()))
```

```python
import numpy as np

category_ids = [np.int64(1), np.int64(2), np.int64(10), np.int64(15), np.int64(17), 
                np.int64(19), np.int64(20), np.int64(22), np.int64(23), np.int64(24), 
                np.int64(25), np.int64(26), np.int64(27), np.int64(28), np.int64(29), 
                np.int64(43)]

# Convert np.int64 to regular integers
category_ids_as_int = [int(i) for i in category_ids]

# Print the list without showing the datatype
print(category_ids_as_int)
```

```python
data.loc[(data['category_id']==1),'category_name'] = 'File And Animation'
data.loc[(data['category_id']==2),'category_name'] = 'Vehicle'
data.loc[(data['category_id']==10),'category_name'] = 'Music'
data.loc[(data['category_id']==15),'category_name'] = 'pets and animals'
data.loc[(data['category_id']==17),'category_name'] = 'sports'
data.loc[(data['category_id']==19),'category_name'] = 'Travel'
data.loc[(data['category_id']==20),'category_name'] = 'Gaming'
data.loc[(data['category_id']==22),'category_name'] = 'People'
data.loc[(data['category_id']==23),'category_name'] = 'Comedy'
data.loc[(data['category_id']==24),'category_name'] = 'Entertainment'
data.loc[(data['category_id']==25),'category_name'] = 'News'
data.loc[(data['category_id']==26),'category_name'] = 'Style'
data.loc[(data['category_id']==27),'category_name'] = 'Education'
data.loc[(data['category_id']==28),'category_name'] = 'Technology'
data.loc[(data['category_id']==29),'category_name'] = 'Non profit'
data.loc[(data['category_id']==43),'category_name'] = 'Shows'
```

```python
data.head()
```

```python
data['year']=data['publish_date'].dt.year
data.head(1)
```

```python
yearly_counts = data.groupby('year')['video_id'].count()

#create a bar chart
yearly_counts.plot(kind='bar', xlabel = 'year' , ylabel = 'Total Publish Count' , title = 'Total Publish Video Per Year')
plp.show()
```

```python
yearly_views = data.groupby('year')['views'].sum()

#create a bar chart
yearly_counts.plot(kind='bar', xlabel = 'year' , ylabel = 'Total views Count' , title = 'Total views Per Year')
plp.xticks(rotation = 0)
plp.show()
```

```python
#sum of views in each category
category_views=data.groupby('category_name')['views'].sum().reset_index()

#Sort the categories by views in descending order
top_categories = category_views.sort_values(by = 'views' , ascending = False ).head(5)

#create a barplot to visualize the top 5 categories
plp.bar(top_categories['category_name'],top_categories['views'])
plp.xlabel('Category Name', fontsize = 12)
plp.ylabel('Total Views',fontsize = 12)
plp.title('Top 5 categories', fontsize =15)
plp.tight_layout()
plp.show()
```

```python
plp.figure(figsize=(12,6))
sn.countplot(x='category_name',data = data, order = data['category_name'].value_counts().index)
plp.xticks(rotation=90)
plp.title('Video Count by Category')
plp.show()
```

```python
#count of number of videos publish per hour
video_per_hour = data['publish_hour'].value_counts().sort_index()

#create a bar plot
plp.figure(figsize=(12,6))
sn.barplot(x = video_per_hour.index, y=video_per_hour.values ,palette = 'rocket')
plp.title('Number of videos published per hour')
plp.xlabel('Hour of Day')
plp.ylabel('Number of Videos')
plp.xticks(rotation = 45)
plp.show()
```

```python
#data['publish_time']=pd.to_datetime(data['publish_time'])
#data['publish_date']=data['publish_time'].dt.date
video_count_by_date = data.groupby('publish_date').size()

plp.figure(figsize=(12,6))
sn.lineplot(data=video_count_by_date)
plp.title('Videos published over time')
plp.xlabel('Publish Date')
plp.ylabel('Number of Videos')
plp.xticks(rotation = 45)

plp.show()
```

```python
#Scatter plot between 'views' and 'Likes'
sn.scatterplot(data=data,x = 'views' , y='likes')
plp.title('Views Vs Likes')
plp.xlabel('Views')
plp.ylabel('Likes')
plp.show()
```
